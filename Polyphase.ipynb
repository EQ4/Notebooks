# Polyphase Interploation

## Support Functions

### `polyize( h, interploation)`

Reorders a vector of FIR filter taps into a polyphase filterbank (PFB).

Say you have 12 filter taps, and want to interpolate by 4. Polyize turns:

    h0, h1, h2, h3, h4, h5, h6, h7, h8, h9, h10, h11

into:

    h0  h1  h2   h3 
    h4  h5  h6   h7 
    h8  h9  h10  h11
    
Each column is a phase of the PFB. In this example, we are interploating by 4, so there are 4 columns. The numbe of taps in each phase depends on how many taps you provide. If the taps are not an integer multiple of interpolation, `polyize` will fill the gaps with 0's.

```julia
function polyize{T}( h::Vector{T}, interpolation )
    hLen         = length( h )
    tapsPerPhase = int( ceil( hLen/interpolation ))
    pfbSize = tapsPerPhase * interpolation
    # check that the vector is an integer multiple of interpolation
    if hLen != pfbSize 
        hExtended             = similar( h, pfbSize )
        hExtended[1:hLen]     = h
        hExtended[hLen+1:end] = 0
        h                     = hExtended
    end
    nFilters     = interpolation
    hLen         = length( h )
    tapsPerPhase = int( hLen/nFilters )
    pfb          = reshape( h, nFilters, tapsPerPhase )'
end
```

### `areApprox( x1, x2 )`

A vectoried version of Base.isapprox. We're going to use this to check that the output of the native interpolator is close enough to IPP's interpolator.

```julia
function areApprox( x1::Vector, x2::Vector )
    Nx1 = length( x1 )
    Nx2 = length( x2 )    
    
    Nx1 == Nx2 || throw( DimensionMismatch("x1 & x2 must be the equal length vectors") )
    
    for i = 1:Nx1
        isapprox( x1[i], x2[i] ) || error( "Something when wrong at index ", i )
    end
end
```

## Native `polyinterpolate( PFB, x)`

```julia
function polyinterpolate{T}( PFB::Array{T}, x::Vector{T} )
    (Ntaps, Nφ) = size( PFB )               # each column is a phase of the PFB, the rows hold the individual taps
    Nx          = length( x )               # number of input items    
    output      = similar( x, Nx * Nφ )     # Noutput = Nx * Nφ
    
    # until Xn == Ntaps, x[Xn-m+1] would reach out of bounds 
    #     this first loop limits Xn-m+1 to a minimum of 1
    #
    # for each input sample, for each φ (phase, or column of the PFB)
    for Xn = 1:Ntaps-1, φ = 1:Nφ                      
        output[Nφ*(Xn-1)+φ] = zero(T)
        # for each tap in phase[n]
        for Tn = 1:Xn
            @inbounds output[Nφ*(Xn-1)+φ] += PFB[Tn, φ] * x[Xn-Tn+1]
        end
    end
    
    # no longer in danger of stepping out of bounds
    #
    # for each input sample, φ (phase, or column of the PFB)
    for Xn = Ntaps:Nx, φ = 1:Nφ
        # for each tap in phase[n]
        output[Nφ*(Xn-1)+φ] = zero(T)
        for Tn = 1:Ntaps
            @inbounds output[Nφ*(Xn-1)+φ] += PFB[Tn, φ] * x[Xn-Tn+1]
        end
    end

    return output
end

polyinterpolate( h, x, interpolation ) = polyinterpolate( polyize( h, interpolation ), x )
```

## SIMD Friendly Version

```julia
function polyinterpolate_v2{T}( PFB::Array{T, 2}, x::Vector{T} )
    (Ntaps, Nφ) = size( PFB )           # each column is a phase of the PFB, the rows hold the individual taps
    Nx          = length( x )           # number of input items
    Ny          = Nx * Nφ
    y           = similar( x, Nx * Nφ ) # Ny = Nx * Nφ
    
    for Xn = 1:Ntaps-1, φ = 1:Nφ        # until Xn == Ntaps, x[Xn-m+1] would reach out of bounds 
                                        # this first loop limits Xn-m+1 to a minimum of 1
        Yn          = Nφ*(Xn-1)+φ
        accumulator = zero(T)

        for Tn = 1:Xn                   # for each tap in phase[n]
            @inbounds accumulator += PFB[Tn, φ] * x[Xn-Tn+1]
        end
        y[Yn] = accumulator
    end
    
    PFB = flipud(PFB)
    
    for Xn = Ntaps:Nx, φ = 1:Nφ         # no longer in danger of stepping out of bounds
        
        XnBase      = Xn-Ntaps               
        Yn          = Nφ*(Xn-1)+φ
        accumulator = zero(T)        
                
        @simd for Tn = 1:Ntaps
            @inbounds accumulator += PFB[Tn, φ] * x[XnBase + Tn]
        end
        
        y[Yn] = accumulator
    end

    return y
end

polyinterpolate_v2( h, x, interpolation ) = polyinterpolate_v2( polyize( h, interpolation ), x )
```

```julia
T  = Float32
Nt = 32
Nx = 100_000
L  = 4

h = IPPDSP.lowpass( T, Nt, 0.5/L, true )
x = rand( T, Nx )

@time ippResult = IPPDSP.filt( h, x, L, 1 )

h = polyize( h, L )

@time polyV1Result = polyinterpolate( h, x )
areApprox( ippResult, polyV1Result )

@time polyV2Result = polyinterpolate_v2( h, x )
areApprox( ippResult, polyV2Result )

df = DataFrame( IPP = ippResult, V1 = polyV1Result, V2 = polyV2Result )
```

## Speed 

```julia
function testspeed( Nsamples, Ntaps, interpolation )

    nTests      = length(Nsamples)*length(Ntaps)
    
    timeData = DataFrame( 
                Samples=DataArray(Int, nTests),
                Taps=DataArray(Int, nTests),
                NativeTime=DataArray(Float64, nTests),
                IPPTime=DataArray(Float64, nTests),
                Relative=DataArray(Float64, nTests) )    

    index      = 0
    
    for nt in Ntaps
    nt = int(nt)

    h = IPPDSP.lowpass( Float64, nt, 0.5/interpolation, true )

        for nx = Nsamples            
            nx = int(nx)
            index += 1

            timeData[index, 1] = nx
            timeData[index, 2] = nt

            nx = int(nx)
            x  = rand( typeof(h[1]), nx )        

            
            t = time()
            gc()
            polyAns = polyinterpolate( h, x, interpolation )
            tPoly = time()-t
            timeData[index, 3] = tPoly
                                
            t = time()
            gc()
            ippAns = IPPDSP.filt( h, x, interpolation, 1 )
            tIPP = time()-t
            timeData[index, 4] = tIPP

            timeData[index, 5] = (tPoly-tIPP)/abs(tIPP)
            
            areApprox( polyAns, ippAns )
        end
    end
    
    return timeData
end
```

```julia
times = testspeed(logspace(2,7, 100), (12, 32, 56, 108), 4);
```

```julia

plot(   times, x="Samples", y="Relative",
        Guide.title("Native Polyphase Interploation"),
        Guide.xlabel("Signal Length"),
        Guide.ylabel("Time Relative to IPP (lower is better)"),
        color="Taps",
        Scale.x_log10,
        Geom.line,
        Scale.discrete_color()
)
```
