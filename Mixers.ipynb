```julia
using Gadfly
using DSP
```

```julia
type Mixer
    sampleRate::Float64
    frequency::Float64
    count::Int64
end
```

```julia
function mix!( buffer::Vector, signal::Vector, sampleRate, frequency, t0 = 0.0 )
    sigLen = length( signal )
    bufLen = length( buffer )
    bufLen >= sigLen || error( "buffer length must be >= signal length" )
    for i = 0:sigLen-1
        t = t0 + i/sampleRate
        buffer[i+1] = signal[i+1] * cos( 2π*frequency*t )
    end
    buffer
end

mix( signal::Vector, sampleRate, frequency, t0 = 0.0 ) = mix!( similar(signal), signal, sampleRate, frequency, t0 )
```

```julia
sampleRate = 100
index = [0:999];
```

```julia
frequency  = 10
x = cos( 2π*frequency*index/sampleRate )
plot(x=linspace(0,99, 100), y=x, Geom.line)
```

```julia
mixFrequency = 10
xMixed = mix( x, sampleRate, mixFrequency);
plot(x=index, y=xMixed, Geom.line)
```

```julia
spectrum = spectrogram( xMixed )
plot(x = spectrum.freq, y = spectrum.power, Geom.point)
```
